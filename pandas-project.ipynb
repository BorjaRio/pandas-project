```python
#analysis of duplicates within serie
df["Case Number"].duplicated().sum()
```

```python
#clean it up and create a list
duplicates = df[df['Case Number'].duplicated(keep=False)]['Case Number'].tolist()
```

```python
display(duplicates)
```

```python
#we do a drop_ duplicates to eliminate duplicates case numbers
data_clean=df.drop_duplicates(subset="Case Number")
```

```python
#doble check for duplicate case numbers
data_clean["Case Number"].duplicated().sum()
```

```python
#We want to know the null numbers. In our case, we just have 6 null numbers in Type of attack.
data_clean.isnull().sum()
```

```python
#We eliminate the null numbers in the column Type
master_data=data_clean.dropna(thresh=3)
display(master_data)
```

```python
#In our analysis, we want to go deeper in the Type. Now we want to know which type of attack was and count it in a list
master_data['Type'].value_counts()
```

```python
#now we want to know which countries have more Case numbers and see the TOP 5
top5_countries= master_data.Country.value_counts().head(5)
top5_countries
```

```python
top5_countries.plot(kind="bar")
plt.xlabel('Country') 
plt.ylabel('Case Number') 
plt.title('Case Number by Country') 
plt.minorticks_on() 
plt.grid() 
```

```python
#now we want to know which activities have more Case numbers and see the TOP 5
top_activity=master_data.Activity.value_counts().head(5)
top_activity
```

```python
top_activity.plot(kind="bar")
plt.xlabel('Activity') 
plt.ylabel('Case Number') 
plt.title('Case Number by Activity') 
plt.minorticks_on() 
plt.grid() 
```

```python
#now we want to know which Areas have more Case numbers and see the TOP 5
top_areas=master_data.Area.value_counts().head(5)
top_areas
```

```python
top_areas.plot(kind="bar")
plt.xlabel('Areas') 
plt.ylabel('Case Number') 
plt.title('Case Number by Areas') 
plt.minorticks_on() 
plt.grid() 
```

```python
top_injury=master_data.Injury.value_counts().head(5)
top_injury
```

```python
top_injury.plot(kind="bar")
plt.xlabel('Injury') 
plt.ylabel('Case Number') 
plt.title('Case Number by Injury') 
plt.minorticks_on() 
plt.grid()
```
